```python
import pandas as pd
```

```python
df=pd.read_csv('customer.csv')
df
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.title().str.replace(' ', '_')
```

```python
df=df.rename(columns = {'Purchase_Amount_(Usd)':'Purchase_Amount'})
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Review_Rating']=df.groupby('Category')['Review_Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.describe(include = 'all')
```

```python
# create column age_group 
import numpy as np
df['Age_group'] = np.where(df['Age']<=13,'Child',np.where(df['Age']<=23,'Teen',np.where(df['Age']<=60,'Adult','Senior')) ) 
```

```python
df[['Age','Age_group']]
```

```python
# PURCHASE Fequency Days 
df['Frequency_Of_Purchases']
```

```python
df['Frequency_Of_Purchases'].replace({'Fortnightly':14,'Weekly':7,'Annually':30,'Quarterly':90,'Bi-Weekly':14,'Monthly':14,'Every 3 Months':90},inplace= True)
```

```python
df['Frequency_Of_Purchases'].unique()
```

```python
df.drop(columns = ['age_group'],axis = 1,inplace = True)
```

```python
# whenever Promo_Code_Used applies Discount_Applied also applied
(df['Discount_Applied'] == df['Promo_Code_Used']).all()
```

```python
#check for  data inconsistency in all columns
df['Item_Purchased'].unique()
df['Gender'].unique()
df['Payment_Method'].unique()
df['Category'].unique()
df['Shipping_Type'].unique()
df['Subscription_Status'].unique()
```
